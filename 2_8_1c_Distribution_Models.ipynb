```python
#1 Normal Distribution
```

```python
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(1234)
```

```python
Samples = np.random.lognormal(mean=1.,sigma=0.4,size=1000)
```

```python
shape,loc,scale = scipy.stats.lognorm.fit(Samples, floc=0)
```

```python
num_bins = 50
```

```python
count,edges,pathconf = plt.hist(Samples,bins=num_bins, color='r')
```

```python
#Binomial distribution
```

```python
import seaborn
from scipy.stats import binom
```

```python
data = binom.rvs(n=20,p=0.5,size=10000)
```

```python
ax = seaborn.histplot(data, kde=True, color='blue', stat='density', binwidth=1, discrete=True, alpha=0.2)
```
