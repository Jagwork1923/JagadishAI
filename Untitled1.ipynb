```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Initialize LabelEncoder
le = LabelEncoder()

# Fit and transform y_train
y_train_encoded = le.fit_transform(y_train)

# Transform y_test using the same encoder fitted on y_train
y_test_encoded = le.transform(y_test)

print("Original y_train labels (first 5):", y_train.head().to_list())
print("Encoded y_train labels (first 5):", y_train_encoded[:5])
print("\nOriginal y_test labels (first 5):", y_test.head().to_list())
print("Encoded y_test labels (first 5):", y_test_encoded[:5])
print("\nClasses encoded:", le.classes_)
```
