```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
```

```python
#1 Load the Dataset
#make sure the filename matches exactly
df = pd.read_csv('FinalData.csv')
```

```python
#2. Inspect the Data
print("---Data Information---")
print(df.info())
print('\n---Missing Values---')
print(df.isnull().sum())
```

```python
#3. Handle Missing Values
# This fills missing numbers with the average (mean)
imputer = SimpleImputer(strategy='mean')
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = imputer.fit_transform(df[numeric_cols])
```

```python
#4.Encode Categorical Data
#Converts text categories into numbers (e.g., 'Yes/no' to 0/1)
le = LabelEncoder()
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
```

### 5. Feature Scaling
Many machine learning algorithms perform better when numerical input variables are scaled to a standard range. Here, we'll use `StandardScaler` to transform our numerical features so they have a mean of 0 and a standard deviation of 1.

```python
scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

print("---Data after scaling (first 5 rows)---")
display(df.head())
```

```python
#5. Split Features (X) and Target (Y)
#Assuming the last column is what you want to predict
X = df.iloc[:, :-1].values
Y = df.iloc[:, -1].values
```

```python
Y
```

```python
#6. Split into Training and Testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=43)
```

```python
X
```

```python
#Feature Scaling
#This puts all numbers on a similar scale(e.g., -1 to 1)
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
print("\npreprocessing Complete!")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
#1. Initialize the model
classifier = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
#Train the model using the scaled training data
classifier.fit(X_train, Y_train)
```

```python
#3. Predict the results for the Test set
y_pred = classifier.predict(X_test)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, confusion_matrix

# Get the unique scaled class values from Y_test
un_scaled_classes = np.unique(Y_test)

# Assume the smaller scaled value maps to class 0 and the larger to class 1
# This relies on the original 'Class' column being 0 and 1, and StandardScaler maintaining order
class_0_scaled = np.min(un_scaled_classes)
class_1_scaled = np.max(un_scaled_classes)

# Create discrete versions of Y_test (true labels)
Y_test_discrete = np.where(Y_test == class_0_scaled, 0, 1)

# For y_pred (continuous predictions), use a threshold to convert to discrete classes
# The midpoint between the two scaled class values is a reasonable threshold
midpoint = (class_0_scaled + class_1_scaled) / 2
y_pred_discrete = np.where(y_pred >= midpoint, 1, 0)

# Now, evaluate the performance using the discrete labels
accuracy = accuracy_score(Y_test_discrete, y_pred_discrete)
print(f"Model Accuracy: {accuracy * 100:.2f}%")
print("\nConfusion Matrix:")
print(confusion_matrix(Y_test_discrete, y_pred_discrete))
```
